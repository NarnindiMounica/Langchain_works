* This notebook is used to build a simple application with LangChain
* Trace application in LangSmith
* Server application with LangServe

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]= "true"
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT")
os.environ["GROQ_API_KEY"]= os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
from langsmith import uuid7
id = uuid7()
model = ChatGroq(model="llama-3.1-8b-instant")
model.invoke("what is machine learning")
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "you are a helpful assistant, answer user queries in friendly tone."),
    ("user", "{input}")
])
prompt
```

```python
chain = prompt | model
response=chain.invoke({"input": " what is machine learning"})
response
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()

chain = prompt | model | parser
result = chain.invoke({"input": "give me 2 lines about agentic ai"})
result
```
