### HuggingFace x LangChain:

langchain_huggingface, a partner package in LangCahin jointly maintained by Hugging Face and LangChain.
This new python package is desgined to bring the powe of the latest development of Huggingface into LangChain and keep it up to date.

#### HuggingFace Endpoint

```python
import os
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from dotenv import load_dotenv
load_dotenv()
```

```python
repo_id="meta-llama/Llama-3.1-8B-Instruct"
llm=HuggingFaceEndpoint(repo_id=repo_id, temperature=0.7, max_new_tokens=150, huggingfacehub_api_token=os.getenv('HF_TOKEN'))
model=ChatHuggingFace(llm=llm)
model

```

```python
model.invoke("say few lines about machine learning").content
```
