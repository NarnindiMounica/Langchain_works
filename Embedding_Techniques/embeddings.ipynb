Embeddings

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "this is an openai embedding sample text"
results = embeddings.embed_query(text)
results
```

```python
# we can set dimensions of embedding vector
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
embeddings_1024
```

Ollama Embeddings

```python
from langchain_ollama import OllamaEmbeddings
ollama_embeddings = OllamaEmbeddings(model="all-minilm")
ollama_embeddings
```

```python
ollama_result = ollama_embeddings.embed_query(text)
ollama_result
```

```python
len(ollama_result)
```
